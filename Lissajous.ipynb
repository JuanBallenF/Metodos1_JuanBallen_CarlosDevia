```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Función que grafica las figuras de Lissajous para cierto desfase.

def figuras_lissajous(delta):
    fig = plt.figure()
    k = 0
    nx = [1,2,3,4,5]
    ny = nx.copy()
    theta = np.linspace(0, 2*np.pi, 100)
    for i in range(5):
        for j in range(5):
            k += 1
            if nx[i] <= ny[j]:
                x = np.sin(nx[i]*theta)
                y = np.sin(ny[j]*theta + delta)
                ax = fig.add_subplot(5,5,k)
                ax.plot(x,y)
                plt.axis('off')
    
```

```python
#Figuras de Lissajous para un desfase de 0.

figuras_lissajous(0)
```

```python
#Figuras de Lissajous para un desfase de π/4.

figuras_lissajous(np.pi/4)
```

```python
#Figuras de Lissajous para un desfase de π/2.

figuras_lissajous(np.pi/2)
```
