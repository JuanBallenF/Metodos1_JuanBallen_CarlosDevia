```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sqrt(np.tan(x))
x = np.linspace(0.1,1.1,100)
fun = f(x)
h = 0.01
```

# Punto C

```python
def ProgressiveDerivative(x,f,h):
    return (f(x+h)-f(x))/h

DerivadaP = ProgressiveDerivative(x,f,h)
```

# Punto D

```python
def CentralDerivative(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)

DerivadaC = CentralDerivative(x,f,h)
```

# Punto E

```python
plt.plot(x, DerivadaC)
plt.plot(x, DerivadaP)
plt.plot(x, (1/(2*((np.cos(x))**2)*(np.sqrt(np.tan(x))))))
```

# Punto F

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)
ax.plot(x, np.abs((1/(2*((np.cos(x))**2)*(np.sqrt(np.tan(x))))) - DerivadaC), label='Derivada Central')
ax.plot(x, np.abs((1/(2*((np.cos(x))**2)*(np.sqrt(np.tan(x))))) - DerivadaP), label='Derivada Progresiva')

fig.legend()
```

Como se ve en la gráfica, ambas aproximaciones tienen precisiones diferentes. La aproximación de la Derivada Central es mucho más precisa, debido a que el orden de magnitud del error es muy pequeño (y muy cercano a cero). Mientras que la derivada progresiva cuenta con un notable aumento en el error, sobre todo en los extremos del intervalo, en estos puntos la derivada progresiva difiere del valor de la derivada calculada analíticamente.
