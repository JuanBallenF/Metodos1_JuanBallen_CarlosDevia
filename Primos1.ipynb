```python
import sympy as sp
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
#(1)
#La función "es_primo" indica si un número es primo o no.
#La función calcula los n primeros números primos y los añade a una lista.

def es_primo(n:int)->bool:
    for i in range (2,n):
        if n%i == 0:
            return False 
    return True

def numeros_primos(n:int)->list:
    primos = []
    i = 2
    while len(primos) < n:
        if es_primo(i) == True:
            primos.append(i)
            i += 1
        else:
            i += 1
    return primos
```

```python
#(1) Lista con los primeros 1000 números primos.
lista = numeros_primos(1000)
lista
```

```python
#(2) Imprimir en pantalla los primeros 10 números primos.

for i in range (1,11):
    print(lista[i])
```

```python
#(3) Gráfica de los primeros 1000 números primos en función de su posición.

plt.plot(range(1,1001), lista, color='r', linewidth=3)
```
