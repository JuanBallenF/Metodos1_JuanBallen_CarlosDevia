```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#GrÃ¡fica de la espiral de arquÃ­medes.

fig = plt.figure()
theta = np.linspace(0, 3*2*np.pi, 1000)
x = theta*np.cos(theta)
y = theta*np.sin(theta)
ax = fig.add_subplot(1,1,1)
ax.plot(x,y)
ax.set_ylabel('y')
ax.set_xlabel('x')
```
