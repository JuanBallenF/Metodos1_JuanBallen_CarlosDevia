```python
import sympy as sp
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#(1) Función que calcula los primeros 20 terminos de la sucesión de Fibonnaci y los devuelve en una lista.

def fibonnaci()->list:
    sucesion = [0, 1]
    i = 2
    for num in range (i,21):
        sucesion.append((sucesion[i-1] + sucesion[i-2]))
        i += 1
    return sucesion
```

```python
#20 primeros términos de la sucesión de Fibonnaci.

fibonnaci()
```

```python
#(2) Gráfica de los primeros 20 términos de la sucesión de Fibonnaci.

lista = fibonnaci()

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(lista, color='k', label='Serie Fibonnaci')
fig.legend()
```

```python
#(3) Función que calcula el número áureo utilizando los términos de la sucesión de Fibonnaci y agrega los resultados 
#a una lista. 

def numero_aureo()->list:
    sucesion = fibonnaci()
    aureo = []
    for i in range(2,21):
        aureo.append(sucesion[i]/sucesion[i-1])
    return aureo
```

```python
#Cálculo del valor exacto del número aureo y gráfica comparandolo con su estimación.

phi = (1 + np.sqrt(5))/2
aureo = numero_aureo()


fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1)
ax2.plot(aureo, label='Estimación usando la serie')
ax2.axhline(phi, color='r', linestyle='--', label='Número Aureo')
fig2.legend()
```
