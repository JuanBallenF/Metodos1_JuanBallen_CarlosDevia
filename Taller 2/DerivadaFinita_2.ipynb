```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 400)
h = 0.05
```

```python
fun = lambda x: 1/np.sqrt(1 + np.exp(-(x**2)))
```

```python
f = fun(x)
```

```python
def DerivativeC (x, f, h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
#Derivada central de la función en el intervalo -10 <= x <= 10

DerivadaC = DerivativeC(x,fun,h)
```

```python
fig = plt.figure(figsize=(5,6))
ax = fig.add_subplot(2,1,1)
ax1 = fig.add_subplot(2,1,2)

ax.plot(x,DerivadaC)
ax.grid()

#Error de la derivada en el intervalo -10 <= x <= 10 con respecto al valor real.

ax1.plot(x, np.abs( ((np.exp(-(x**2))*x)/((1 + np.exp(-(x**2)))**(3/2))) - DerivadaC))
ax1.grid()
```
