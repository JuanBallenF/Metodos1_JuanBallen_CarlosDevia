```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

$$ {M} = [1, -2, 1] $$


$$ {f''}(x) = \frac{{f}(x+h) - 2{f}(x) + {f}(x-h)}  {h^2} $$

```python
x = np.linspace(-10, 10, 400)
m = np.array([1,-2,1])
h = 0.05
```

```python
fun = lambda x: 1/np.sqrt(1 + np.exp(-(x**2)))
```

```python
f = fun(x)
```

```python
def SecondDerivative (x, f, h, m):
    der = np.array([])
    for i in range(len(x)):
        if i == 0 or i == (len(x)-1):
            der = np.append(der, 0)
        else:
            der = np.append(der, (((f[i-1]*m[0]) + (f[i]*m[1]) + (f[i+1]*m[2]))/(h*h)))
                            
        i += 1
    return der
```

```python
SegundaDerivada = SecondDerivative(x, f, h, m)
```

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)


ax.plot(x,SegundaDerivada)
ax.grid()
```
