```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def coefficients(N, k):
    
    x = np.random.uniform(size=int(N+k))
    sum_ = 0
    for i in range(N):
        
        sum_ += x[i]*x[i+k]
        
    return (1/N)*sum_
```

```python
N = int(10e4)
k = 30
coeffs = []
for i in range(1,k+1):
    coeffs.append(coefficients(N,i))
    
plt.plot(range(1,k+1), coeffs)
plt.ylim(0.20,0.35)
plt.ylabel('C(k)')
plt.xlabel('k-esimo vecino')
plt.grid()
```
