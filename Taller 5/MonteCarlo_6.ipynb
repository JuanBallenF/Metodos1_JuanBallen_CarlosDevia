```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import integrate
import sympy as sym
```

$$ \int\int\int e^{ \sqrt{ x^2 + y^2 + z^2 } } dxdydz  $$
$$ \{x^2 + y^2 + z^2 \leq 1\}: $$

```python

phi = np.random.uniform(0,np.pi,size=int(100000))
theta = ( np.random.uniform(0,2*np.pi,size=int(100000)) )
r = (np.random.uniform(0,1,size=int(100000)))


x_values = ( r * np.sin( phi) * np.cos( theta ) )
y_values = ( r * np.sin( theta) * np.sin( phi ) )
z_values = ( r * np.cos( phi ) )
```

```python
def f(x,y,z):
    return np.e**(np.sqrt((x**2) + (y**2) + (z**2)))

def MCIntegral( x,y,z):
    
    a = []
    
    vol = 4*np.pi/3

    for i in range(len(x)):
         
        a.append(f(x[i],y[i],z[i]))   

    return np.mean(a)*vol
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection = '3d')

ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)


ax.scatter(x_values,y_values,z_values, color='b')
```
