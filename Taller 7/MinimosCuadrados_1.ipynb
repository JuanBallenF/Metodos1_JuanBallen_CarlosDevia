```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from matplotlib import cm
```

```python
def r_1(x):
    return 2*x -2

def r_2(x):
    return (1 - x)/2

def r_3(x):
    return 4 - x
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
b = np.array([[2],[1],[4]])
print(A)
print(b)
```

# a)

```python

def GetFit(x,y):
 
    b = y

    AT = np.dot(x.T,x)
    bT = x.T @ b

    xsol = np.linalg.solve(AT,bT)
    
    return xsol
```

```python
P = GetFit(A,b)
P
```

```python
x_ = np.linspace(-10,10,100)
y_1 = []
y_2 = []
y_3 = []

for i in range(len(x_)):
    y_1.append(r_1(x_[i]))
    y_2.append(r_2(x_[i]))
    y_3.append(r_3(x_[i]))

plt.plot(x_,y_1)
plt.plot(x_,y_2)
plt.plot(x_,y_3)
plt.scatter(P[0],P[1],color='r',zorder=3)
```

```python
np.linalg.norm(np.dot(A,P) - b)
```

El vector x sería la solución aproximada al sistema Ax = b. Equivale al punto que más se aproxima a la intersección de las rectas, debido a que estas tres rectas no se intersectan en ningún punto. \
Ax equivale a la proyección del vector b sobre el espacio columna de A, por tanto, al calcular min( || Ax - b || ) se tiene el valor de x que minimiza la distancia entre la proyección y b, por tanto, es la solución más cercana al sistema.

# b)

```python

x = np.linspace(-5,5,int(300))
y = np.linspace(-5,5,int(300))
```

```python
X, Y = np.meshgrid(x, y)

Distance = np.array([])

for i in range(len(X)):
    
    for j in range(len(Y)):
        
        d = np.linalg.norm(np.dot(A,[[x[i]],[y[j]]]) - b)
        Distance = np.append(Distance,d)
        
```

```python
Z= Distance.reshape(300,300)
np.min(Z)
```

```python

fig_1 = plt.figure(figsize=(6,6))

ax1 = fig_1.add_subplot(1,1,1, projection='3d')

ax1.plot_surface(Y,X,Z,cmap=cm.coolwarm,linewidth=0, antialiased=False)

ax1.set_ylim(5,-5)
ax1.set_xlim(-5,5)
ax1.set_ylabel('X')
ax1.set_xlabel('Y')
ax1.set_zlabel('||Ax − b||')
```
