```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

$$ P(CCSS) = P_1P_2P_3P_4 + P_1(1-P_2)P_3P_4 + (1-P_1)P_2P_3P_4 + P_3P_4 $$
$$\ $$
$$ P(CCSS) = XY\left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + X(1-Y)\left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + (1-X)Y\left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + \left(\frac{1}{2}\right)\left(\frac{1}{2}\right) $$
$$\ $$
$$ P(CCSS) = \frac{X+Y}{4} - \frac{XY}{2} + \frac{1}{4} $$

```python

P1 = np.linspace(0.1,0.9,100)
P2 = np.linspace(0.1,0.5,100)

X, Y = np.meshgrid(P1,P2)

Z = ( X*(1-Y)*(1/2)*(1/2) ) + ( Y*(1-X)*(1/2)*(1/2) ) + (1/2)*(1/2)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X,Y,Z,cmap=cm.coolwarm,linewidth=0, antialiased=True)

ax.set_xlabel('P1', fontsize=12)
ax.set_ylabel('P2', fontsize=12)
ax.set_zlabel('Pt', fontsize=12)
ax.view_init(20,-140)
```

```python
Z_list = []
a = []

for i in range(len(P1)):
    
    for j in range(len(P2)):
        
        z = ( P1[i]*(1-P2[j])*(1/2)*(1/2) ) + ( P2[j]*(1-P1[i])*(1/2)*(1/2) ) + (1/2)*(1/2)
        Z_list.append((z,P1[i],P2[j]))
        a.append(z)
    
```

# Probabilidad Mínima:

```python
np.argmin(a)
```
